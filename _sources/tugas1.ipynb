## **1. Membaca Data Dari URL**

```python
#Import Dataset
import pandas as pd
url = "https://raw.githubusercontent.com/KharismaIntanSafitri/datamining/main/IRIS.csv"
dataset = pd.read_csv(url)
```

```python
dataset.info()
```

```python
dataset
```

## **2. Membuat statistik dari data setiap atribut yang bertipe**

### **2.1 Rata-rata (Mean)**

```python
dataset["sepal_length"].mean()
```

### **2.2 Nilai sering muncul (Modulus)**

```python
dataset["sepal_length"].mode()
```

### **2.3 Nilai Tengah (Median)**

```python
dataset["sepal_length"].median()
```

## **3. Mendeskripsikan kemiringan / skwened dari data setiap variabel**

```python
dataset["sepal_length"].skew()
```
